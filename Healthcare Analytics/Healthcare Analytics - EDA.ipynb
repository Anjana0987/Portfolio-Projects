### Lets look at the beneficiary data

```python
display(train_beneficiary.head())
print("No of distinct members: ", train_beneficiary.BeneID.nunique())
```

```python
# Change the value form 2 to 0 to make them binary columns
chronic_conditions_columns = [i for i in train_beneficiary.columns if 'ChronicCond' in i]

for i in chronic_conditions_columns:
    train_beneficiary[i] = np.where(train_beneficiary[i] == 2, 0, train_beneficiary[i])

train_beneficiary.head()
```

```python
# Convert DOB and DOD to datetime cols
train_beneficiary["DOB"] = pd.to_datetime(train_beneficiary["DOB"])
train_beneficiary["DOD"] = pd.to_datetime(train_beneficiary["DOD"])

# Also create a is_dead column
train_beneficiary["is_dead"] = np.where(train_beneficiary.DOD.notnull(), 1, 0)

train_beneficiary.head()
```
